# Import Necessary Libraries

```python
# Question "Can we compare the diets and their environmental impacts across all categories?"

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
```

# Read data

```python
data = pd.read_csv("Results_21Mar2022.csv")  # read the data
```

# Check for duplicate rows, missing values and unique categories

```python

print("Shape of the original data:", data.shape)  # let's see our data's shape

missing_values = data.isnull().sum()  # check for missing values
print("\nMissing values:\n", missing_values)  # our data has no missing values

duplicates = data.duplicated().sum()  # check for duplicates
print("\nNumber of duplicate rows:\n", duplicates)  # our data has no duplicate rows

# let's check for unique values in diet_group, age_group and grouping
# we check to ensure labels are uniform  (they are, it tuns out)
print(f"unique values in data_group: {data['diet_group'].unique()}")
print(f"unique values in age_group: {data['age_group'].unique()}")
print(f"unique values in grouping: {data['grouping'].unique()}")
```

# Reducing Complexity of the Data

To reduce complexity, we aggregate the data
And we group by diet_group and sex because we want to compare the diets against environmental impacts
We will drop grouping and age group because the age doesn't matter for our inquiry.

```python
relevant_data = data.drop(['grouping', 'age_group'], axis=1)
aggregated_data = relevant_data.groupby(['diet_group', 'sex']).mean().reset_index()
print("\nShape of the aggregated data after removing duplicates:", aggregated_data.shape)

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
# print(data)
aggregated_data
```

The new shape is (12, 22) which is much simpler to visualize. We've traded a detailed, granular view to a summarized, high-level one.

Now, we are going to normalise our data so that features are the same scale and therefore more comparable.

lets first see the distribution of each feature to get a good idea of what scaling technique to use

```python
aggregated_data.drop('mc_run_id', axis=1, inplace=True)
plt.figure(figsize=(10,6))
for i, column in enumerate(aggregated_data.columns[2:]):  # exclude the categorical colums
    # print(column)
    plt.subplot(4, 5, i + 1)
    sns.histplot(aggregated_data[column], kde=True)  # represent as a smooth curve
    plt.title(column)

plt.tight_layout()
plt.show()
```

Considering that quite a number of features show skewness due to outliers, lets use RobustScaler because its robust to outliers

```python
scaler = RobustScaler()
scaled_data = scaler.fit_transform(aggregated_data.iloc[:, 2:])  # fit and transform data while excluding categories
scaled_df = pd.DataFrame(scaled_data, columns=aggregated_data.columns[2:])  # convert scaled data back to a dataframe
scaled_df = pd.concat([aggregated_data.iloc[:, :2], scaled_df], axis=1)  # concatenate the categorical columns

print("\nData frame after scaling:\n") 
scaled_df

# print(scaled_df.dtypes)
```

# Visualizing the data

Before we visualize the data, lets group the data by sex and save that into two new dataframes so that we can compare diets across different environmental factors and see if any difference in the impacts exists for males compared to females.

```python
grouped_by_sex = scaled_df.groupby('sex')

males = grouped_by_sex.get_group('male').copy()
males.drop('sex', axis=1, inplace=True)

females = grouped_by_sex.get_group('female').copy()
females.drop('sex', axis=1, inplace=True)

df_list = {"Females": females, "Males": males}
```

Because there are so many variables that make up environmental factors, we will use a parallel coordinate plot as that's great for visualizing multivariat data.

```python
fig, axis = plt.subplots(2, 1, figsize=(15, 12), sharey=True)

for i, (key, df) in enumerate(df_list.items()):
    pd.plotting.parallel_coordinates(df, 'diet_group', ax=axis[i], colormap='viridis')
    axis[i].set_title(f"PCP of Environmental Metrics by Diet Group for {key}")
    axis[i].set_xlabel('Environmental Metric')
    axis[i].set_ylabel('Normalized Value')
    axis[i].tick_params(axis='x', rotation=45)

plt.suptitle('Parallel Coordinate Plot (PCP) of Environmental Metrics by Diet Group for Different Genders')
plt.legend(loc='best')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
